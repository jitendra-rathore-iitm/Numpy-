```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("************Zomato Sales analysis**************")
print("\n Sales Data: \n ",sales_data)
print("\n Shape of this data: ", sales_data.shape)
print(" \n Sample data for the first 3 restaurant: \n", sales_data[0:3])
```

```python
# total sales per year
print("Total sales per year: ",np.sum(sales_data[:, 1:], axis = 0))
```

```python
#Total sales for Burger Point
print("Total Sales for Burger Point: ", np.sum(sales_data[3:4, 1:], axis = 1))
```

```python
# Minimum Sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis = 1)
print("Minimum Sales per restaurant: ", min_sales)
```

```python
#Minimum Sales per year
min_sales = np.min(sales_data[:, 1:], axis = 0)
print("Minimum Sales per year: ", min_sales)
```

```python
#Maximum sales per restaurant
max_sales = np.max(sales_data[:, 1:], axis = 1)
print("Maximum Sales per restaurant: ",max_sales)
```

```python
# Maximum Sales per year
max_sales = np.max(sales_data[:, 1:], axis = 0)
print("Maximum Sales per year: ", max_sales )
```

```python
#Average Sales Per year
avg_sales = np.mean(sales_data[:, 1:], axis = 0)
print("Average Sales per year: ", avg_sales)
```

```python
#Average Sales Per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis = 1)
print("Average Sales per restaurant: ", avg_sales)
```

```python
#Cumulative sales per restaurant
cumsum = np.cumsum(sales_data[:, 1:], axis = 1)
print("Cumulative Sales Per restaurant: \n", cumsum)
```

```python
plt.figure(figsize = (10, 6))
plt.plot(np.mean(cumsum, axis = 0))
plt.title("Average Cumulative Sales across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
