```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.random.random((3, 4))
arr3 = np.full((3, 4), 9)
```

```python
np.save('array1.npy', arr1)
np.save('array2.npy', arr2)
np.save('array3.npy', arr3)
```

```python
np.load('array1.npy')
```

```python
try:
    load = np.load('numpy-logo.npy')

    #Display
    plt.figure(figsize = (10, 5))
    plt.subplot(121)
    plt.imshow(load)
    plt.grid(False)
    plt.title("Numpy Logo")

    #Dark Numpy logo
    plt.figure(figsize = (10, 5))
    plt.subplot(121)
    plt.imshow(1-load)
    plt.grid(False)
    plt.title("Dark Logo")

except:
    print("Something Went Wrong :)")
```
